# 거리두기 확인하기

https://programmers.co.kr/learn/courses/30/lessons/81302?language=python3

```python
places = [
    ["POOOP", "OXXOX", "OPXPX", "OOXOX", "POXXP"], 
    ["POOPX", "OXPXP", "PXXXO", "OXXXO", "OOOPP"], 
    ["PXOPX", "OXOXP", "OXPOX", "OXXOP", "PXPOX"], 
    ["OOOXX", "XOOOX", "OOOXX", "OXOOX", "OOOOO"], 
    ["PXPXP", "XPXPX", "PXPXP", "XPXPX", "PXPXP"]
]

def solution(places):    
    def check(x, y, place, n):
        dx = [-1, 1, 0, 0]
        dy = [0, 0, -1, 1]
        threshold = 0
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            if -1 < nx < 5 and -1 < ny < 5 and place[nx][ny] == 'P':
                 threshold += 1
            if threshold >= n:  # 거리두기 실패
                return True 
        return False         

    res = []
    for place in places:
        sb = False
        for x in range(5):
            for y in range(5):
                if place[x][y] == 'O' and check(x, y, place, 2) or place[x][y] == 'P' and check(x, y, place, 1): 
                    '''
                    위의 조건이 걸린다면 거리두기를 지키지 않는 것.
                    1. 책상을 중심으로 상하좌우에 사람이 2명 이상 있으면 거리두기 X
                    2. 사람을 중심으로 상하좌우에 사람이 1명이라도 있다면 거리두기 X
                    '''			              
                    res.append(0)
                    sb = True  # 조건문 2번 탈출을 위해
                    break
            if sb:
                break
        else:  # for문을 끝까지 잘 돌고 끝났을 경우
            res.append(1)
    return res

print(solution(places))
```
