```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
# 새로운 컬럼 생성
df['Age_0'] = 0
df.head()
```

```python
df.drop('Age_0', axis=1, inplace=True)
df.head()
```

```python
# 정렬
df.sort_values(by='Name', ascending=False)
```

```python
# groupby
df.groupby('Pclass')['Age'].agg([max, min])
```

```python
# 컬럼별로 다른 방식으로 묶기
agg_format = {'Age':'max', 'SibSp':'sum', 'Fare':'mean'}
df.groupby('Pclass').agg(agg_format)
```
