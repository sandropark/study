## 캐릭터 1개만 사용

```python
x_data = np.array([[[1,0,0,0]]], dtype=np.float32)

hidden_size = 2
cell = tf.keras.layers.SimpleRNNCell(units=hidden_size)
rnn = tf.keras.layers.RNN(cell, return_sequences=True, return_state=True)
outputs, _states = rnn(x_data)

print(f"x_data: {x_data}, shape: {x_data.shape}")
print(f"outputs: {outputs}, shape: {outputs.shape}")    # 전체 시퀀스에 해당하는 스테이트값
print(f"states: {_states}, shape: {_states.shape}")       # 시퀀스의 마지막 스테이트값
```

# ML lab12-2: RNN - Hi Hello Training

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras

idx2char = ['h', 'i', 'e', 'l', 'o']
y_data = [[1, 0, 2, 3, 3, 4]]  # ihello

hidden_size = 5 
input_dim = 5  # 캐릭터의 수가 5개다. 
sequence_length = 6  # 입력하는 단어 hihell가 6개이기 때문|ihello| == 6 
batch_size = 1
learning_rate = 0.1

x_one_hot = np.array([[[1, 0, 0, 0, 0],    # h 0
                       [0, 1, 0, 0, 0],    # i 1
                       [1, 0, 0, 0, 0],    # h 0
                       [0, 0, 1, 0, 0],    # e 2
                       [0, 0, 0, 1, 0],    # l 3
                       [0, 0, 0, 1, 0]]],  # l 3
                     dtype=np.float32)

y_one_hot = keras.utils.to_categorical(y_data, num_classes=hidden_size)
```

```python
model = keras.Sequential()

 # 유닛은 5다. (원핫인코딩한 벡터로 출력하기 때문), 인풋 shape은 먼저 총 입력할 데이터수 (6)개, 그리고 각 데이터의 차원 (5개의 원핫인코딩 벡터)
cell = keras.layers.LSTMCell(units=hidden_size, input_shape=(sequence_length, input_dim))  
model.add(keras.layers.RNN(cell=cell, return_sequences=True))
model.add(keras.layers.TimeDistributed(keras.layers.Dense(units=hidden_size, activation='softmax')))

model.compile(loss='categorical_crossentropy', 
              optimizer=keras.optimizers.Adam(learning_rate=learning_rate), 
              metrics=['accuracy'])
```

```python
model.fit(x_one_hot, y_one_hot, epochs=50, batch_size=batch_size)
model.summary()
```

```python
predictions = model.predict(x_one_hot)
for i, prediction in enumerate(predictions):
    print(prediction)
    result_str = [idx2char[c] for c in np.argmax(prediction, axis=1)]
    print('\nPrediction:', np.argmax(prediction, axis=1))
    print("Prediction str:", ''.join(result_str))
```

# ML lab12-3: Long Sequence RNN


```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

```python
sample = 'if you want you'

idx2char = list(set(sample))    # 인덱스를 넣으면 캐릭터를 뱉는다.
# idx2char : ['y', 'u', 'a', 'n', 'w', 'i', 't', 'o', 'f', ' ']

char2idx = {c: i for i,c in enumerate(idx2char)}    # 캐릭터를 넣으면 인덱스를 뱉는다.
# char2idx : {' ': 9, 'a': 2, 'f': 8, 'i': 5, 'n': 3, 'o': 7, 't': 6, 'u': 1, 'w': 4, 'y': 0} 

sample_idx = [char2idx[c] for c in sample]  # 샘플을 인덱스로 변경
# sample_idx [5, 8, 9, 0, 7, 1, 9, 4, 2, 3, 6, 9, 0, 7, 1]

x_data = [sample_idx[:-1]]  # 'if you want yo'
y_data = [sample_idx[1:]]   # 'f you want you'
```

```python
# 하이퍼 파라미터 
num_classes = len(idx2char)         # 10
dic_size = len(char2idx)            # 10
rnn_hidden_size = len(char2idx)     # 10
sequense_length = len(sample) - 1   # 14
batch_size = 1
learning_rate = 0.1
```

```python
x_one_hot = tf.one_hot(x_data, num_classes) # (1, 14, 10)
y_one_hot = tf.one_hot(y_data, num_classes) # (1, 14, 10)
```

```python
model = keras.Sequential();
model.add(keras.layers.LSTM(units=num_classes, 
                            input_shape=(x_one_hot.shape[1], x_one_hot.shape[2]), 
                            return_sequences=True))
model.add(keras.layers.TimeDistributed(keras.layers.Dense(units=num_classes, activation='softmax')))
model.summary()

model.compile(loss='categorical_crossentropy', 
              optimizer=keras.optimizers.Adam(learning_rate=learning_rate),
              metrics=['accuracy'])

model.fit(x_one_hot, y_one_hot, epochs=50)
```

```python
predictions = model.predict(x_one_hot)

for i, prediction in enumerate(predictions):
    result_str = [idx2char[c] for c in np.argmax(prediction, axis=1)]
    print("Prediction str: ", ''.join(result_str))
```

## 긴 문자열 사용하기

```python
sentence = ("if you want to build a ship, don't drum up people together to "
            "collect wood and don't assign them tasks and work, but rather "
            "teach them to long for the endless immensity of the sea.")

char_set = list(set(sentence))                     # 인덱스를 넣으면 캐릭터를 뱉는다.
char_dic  = {c: i for i,c in enumerate(char_set)}   # 캐릭터를 넣으면 인덱스를 뱉는다.
```

```python
data_dim = len(char_set)    # 25
hidden_size = len(char_set) # 25
num_classes = len(char_set) # 25
sequence_length = 10  # Any arbitrary number
learning_rate = 0.1
```

```python
dataX = []
dataY = []
for i in range(len(sentence) - sequence_length):
    x_str = sentence[i:i + sequence_length]
    y_str = sentence[i + 1: i + sequence_length + 1]
    # print(i, x_str, '->' , y_str)

    # 인덱스로 변경
    x = [char_dic[c] for c in x_str]
    y = [char_dic[c] for c in y_str]

    dataX.append(x)
    dataY.append(y)

batch_size = len(dataX)
```

```python
X_one_hot = tf.one_hot(dataX, num_classes)  # (170, 10, 25)
Y_one_hot = tf.one_hot(dataY, num_classes)  
```

```python
model = keras.Sequential()

model.add(keras.layers.LSTM(units=num_classes, input_shape=(sequence_length, X_one_hot.shape[2]), return_sequences=True))
model.add(keras.layers.LSTM(units=num_classes, return_sequences=True))
model.add(keras.layers.TimeDistributed(keras.layers.Dense(units=num_classes, activation='softmax')))
model.summary()

model.compile(loss='categorical_crossentropy', 
              optimizer=keras.optimizers.Adam(learning_rate=learning_rate),
              metrics=['accuracy'])

model.fit(X_one_hot, Y_one_hot, epochs=100)
```

```python
results = model.predict(X_one_hot)
for j, result in enumerate(results):
    index = np.argmax(result, axis=1)
    if j == 0:  # print all for the first result to make a sentence
        print(''.join([char_set[t] for t in index]), end='')
    else:
        print(char_set[index[-1]], end='')
```
