# Logistic regression 구현

```python
import tensorflow as tf
import numpy as np

X = np.array([[1, 2],
            [2, 3],
            [3, 1],
            [4, 3],
            [5, 3],
            [6, 2]], dtype='float32')
Y = np.array([[0],[0],[0],[1],[1],[1]],dtype='float32')

W = tf.Variable(tf.random.normal([2, 1]), name='weight')
b = tf.Variable(tf.random.normal([1]), name='bias')
```

```python
# hypothesis

# hypothesis = tf.math.divide(1., 1. + tf.exp(tf.matmul(X, W) + b))
hypothesis = tf.sigmoid(tf.matmul(X, W) + b)
hypothesis
```

```python
# cost function
cost = -tf.reduce_mean(Y * tf.math.log(hypothesis) + 
                       (1 - Y) * tf.math.log(1 - hypothesis))
cost
```

```python
# train

learning_rate = 0.01
for i in range(5000):
    with tf.GradientTape() as tape:
        hypothesis = tf.sigmoid(tf.matmul(X, W) + b)
        cost = -tf.reduce_mean(Y * tf.math.log(hypothesis) \
                + (1 - Y) * tf.math.log(1 - hypothesis))

    W_grad, b_grad = tape.gradient(cost, [W, b])
    
    W.assign_sub(learning_rate * W_grad)
    b.assign_sub(learning_rate * b_grad)
```

```python
# prediction 

predicted = tf.cast(hypothesis > 0.5, dtype=tf.float32)
predicted
```

```python
# accuracy

accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype=tf.float32))
accuracy
```

## 전체 코드

```python
import tensorflow as tf
import numpy as np

X = np.array([[1, 2],[2, 3],[3, 1],[4, 3],[5, 3],[6, 2]], dtype='float32')
Y = np.array([[0],[0],[0],[1],[1],[1]],dtype='float32')

W = tf.Variable(tf.random.normal([2, 1]), name='weight')
b = tf.Variable(tf.random.normal([1]), name='bias')

epochs = 10001
learning_rate = 0.01

for step in range(epochs):
    with tf.GradientTape() as tape:
        hypothesis = tf.sigmoid(tf.matmul(X, W) + b)
        cost = -tf.reduce_mean(Y * tf.math.log(hypothesis) \
                + (1 - Y) * tf.math.log(1 - hypothesis))

    W_grad, b_grad = tape.gradient(cost, [W, b])
    
    W.assign_sub(learning_rate * W_grad)
    b.assign_sub(learning_rate * b_grad)

    if step % 500 == 0:
        print(step, cost)

predicted = tf.cast(hypothesis > 0.5, dtype=tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype=tf.float32))

print('\n',hypothesis,'\n\n', predicted,'\n\n',accuracy)
```

## 모델 사용하기

```python
import pandas as pd

data = pd.read_csv('https://raw.githubusercontent.com/hunkim/DeepLearningZeroToAll/master/data-03-diabetes.csv',header=None)
data.head()
```

```python
import numpy as np
import tensorflow as tf

xy = np.loadtxt('https://raw.githubusercontent.com/hunkim/DeepLearningZeroToAll/master/data-03-diabetes.csv',delimiter=',', dtype=np.float32)
x_data = xy[:, :-1]
y_data = xy[:, -1]
```

```python
# 1. 모델 객체 생성 및 레이어 추가
tf.model = tf.keras.Sequential()
tf.model.add(tf.keras.layers.Dense(units=1,input_dim=x_data.shape[1], activation='sigmoid'))

# 2. 모델 컴파일 (loss, optimizer, metrics 지정)
tf.model.compile(
    loss='binary_crossentropy', 
    optimizer=tf.keras.optimizers.SGD(lr=0.01), 
    metrics=['accuracy']
)
tf.model.summary()

# 3. 훈련
history = tf.model.fit(x_data, y_data, epochs=500)
```

```python
# accuracy
print(f"Accuracy: {history.history['accuracy'][-1]}")

# predict a single data point
y_predict = tf.model.predict([[0.176471, 0.155779, 0, 0, 0, 0.052161, -0.952178, -0.733333]])
print(f"Predict: {y_predict}")

# evaluating model
evaluate = tf.model.evaluate(x_data, y_data)
print((f"loss: {evaluate[0]}, Accuracy: {evaluate[1]}"))
```

## 모델 사용하지 않고 직접 구현하기

```python
import numpy as np
import tensorflow as tf

xy = np.loadtxt('https://raw.githubusercontent.com/hunkim/DeepLearningZeroToAll/master/data-03-diabetes.csv',delimiter=',', dtype=np.float32)
X = xy[:, :-1]
Y = xy[:, -1]
```

```python
# W, b 선언

W = tf.Variable(tf.random.normal([X.shape[1], 1]), name='weight')
b = tf.Variable(tf.random.normal([1]), name='bias')
```

```python
epochs = 5001
learning_rate = 0.01
for step in range(epochs):
    with tf.GradientTape() as tape:
        hypothesis = tf.sigmoid(tf.matmul(X, W) + b)
        cost = -tf.reduce_mean(Y * tf.math.log(hypothesis) + (1 - Y) * tf.math.log(1-hypothesis))
    W_grad, b_grad = tape.gradient(cost, [W, b])
    W.assign_sub(learning_rate * W_grad)
    b.assign_sub(learning_rate * b_grad)
    if step % 250 == 0:
        print(step, cost.numpy())
```

```python
predicted = tf.cast(hypothesis > 0.5, dtype=tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype=tf.float32))

print(f"Hypothesis: {hypothesis[:5]}, \n\n predicted: {predicted[:5]}, \n\n accuracy: {accuracy}")
```
