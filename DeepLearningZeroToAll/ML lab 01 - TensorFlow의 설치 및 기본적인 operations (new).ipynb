```python
import tensorflow as tf

# import tensorflow.compat.v1 as tf
# tf.disable_v2_behavior()
```

```python
hello = tf.constant('Hello, TensorFlow!')

print(hello)
```

```python
node1 = tf.constant(3.0)
node2 = tf.constant(4.0)
node3 = tf.add(node1, node2)

print(node3)
print(node1 + node2)
```

Function

```python
@tf.function
def adder(a, b):
    return a + b

print(adder(3, 5))

A = tf.constant([1,3])
B = tf.constant([2,4])
print(adder(A, B))
```
